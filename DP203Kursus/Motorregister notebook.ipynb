```

dfSelected = df.select('ns:KoeretoejArtNavn','ns:RegistreringNummerNummer','ns:KoeretoejOplysningGrundStruktur')
```

```
from pyspark.sql.functions import *
display(dfSelected.select('ns:KoeretoejOplysningGrundStruktur.ns:KoeretoejBetegnelseStruktur.*'))
```

```
dfKoeretoej=dfSelected.where(col('ns:KoeretoejArtNavn')=='Personbil').select('ns:KoeretoejArtNavn','ns:RegistreringNummerNummer','ns:KoeretoejOplysningGrundStruktur.ns:KoeretoejBetegnelseStruktur.ns:KoeretoejMaerkeTypeNavn','ns:KoeretoejOplysningGrundStruktur.ns:KoeretoejBetegnelseStruktur.ns:Model.ns:KoeretoejModelTypeNavn','ns:KoeretoejOplysningGrundStruktur.ns:KoeretoejOplysningFoersteRegistreringDato') 
display(dfKoeretoej)
```

```
dfData=(dfKoeretoej.select(col('ns:KoeretoejArtNavn').alias('Type')
                           ,col('ns:RegistreringNummerNummer').alias('RegistreringsNr')
                           ,col('ns:KoeretoejMaerkeTypeNavn').alias('Maerke')
                           ,col('ns:KoeretoejModelTypeNavn').alias('Model')
                           ,col('ns:KoeretoejOplysningFoersteRegistreringDato').alias('RegDato')))
```

```
permanent_table_name='DatabricksDB.motordata2'
dfData.write.format("parquet").saveAsTable(permanent_table_name)
```
