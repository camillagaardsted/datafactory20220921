```
# se  https://docs.microsoft.com/en-us/azure/databricks/security/credential-passthrough/adls-passthrough#--mount-azure-data-lake-storage-to-dbfs-using-credential-passthrough
configs = {
  "fs.azure.account.auth.type": "CustomAccessToken",
  "fs.azure.account.custom.token.provider.class": spark.conf.get("spark.databricks.passthrough.adls.gen2.tokenProviderClassName")
}

# Optionally, you can add <directory-name> to the source URI of your mount point.
dbutils.fs.mount(
  source = "abfss://data@datalake20220915.dfs.core.windows.net/",
  mount_point = "/mnt/data",
  extra_configs = configs)
```

```
%fs ls /mnt/data
```

```
%sql

-- https://docs.databricks.com/data/data-sources/xml.html


CREATE TABLE books
USING xml
OPTIONS (path "dbfs:/mnt/data/books.xml", rowTag "book")
```

```
%sql
DESCRIBE TABLE EXTENDED books
```
